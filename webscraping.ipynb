```python
import bs4 as bs
import urllib.request
```

```python
sauce = urllib.request.urlopen('https://www.imdb.com/india/top-rated-indian-movies/?ref_=nv_mv_250_in').read()
```

```python
sauce
```

```python
soup = bs.BeautifulSoup(sauce,'lxml')
soup
```

```python
soup.title
```

```python
soup.title.name
```

```python
soup.title.string
```

```python
soup.title.text
```

```python
body = soup.body
for paragraph in body.find_all('p'):
    print(paragraph.text)
```

```python
for div in soup.find_all('div'):
    print(div.text)
```

```python
table = soup.table
table
```

```python
table_rows = table.find_all('tr')
for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    print(row)
```

```python
import pandas as pd
dfs = pd.read_html('https://www.imdb.com/india/top-rated-indian-movies/?ref_=nv_mv_250_in',header=0)
for df in dfs:
    print(df)
```
